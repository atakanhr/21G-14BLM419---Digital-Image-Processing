```python
def get_value_from_triple(img):
    return int(img[0]/3 + img[1]/3 + img[2]/3)
```

```python
def get_0_1_from_triple(img):
    temp = int(img[0]/3 + img[1]/3 + img[2]/3)
    
    val = 0 if temp < 110 else 1
    
    return val
```

```python
def convert_rgb_to_gray(im_1):
    """RGB  -> Gray scale (0,255) """
    m,n,k = im_1.shape

    new_img = np.zeros((m,n),dtype = 'uint8')

    for i in range(m):

        for j in range(n):

            s = get_value_from_triple(im_1[i,j,:])
            
            new_img[i,j] = s
    
    return new_img
```

```python
def convert_rgb_to_bw(im_1):
    m,n,k = im_1.shape
    
    new_img = np.zeros((m,n),dtype='uint8')
    
    for i in range(m):

        for j in range(n):

            s = get_0_1_from_triple(im_1[i,j,:])
            
            new_img[i,j] = s
    
    return new_img
```

```python
gray_img = convert_rgb_to_gray(im_1)
```

```python
plt.imshow(gray_img,cmap='gray')
```

```python
black_and_white_img = convert_rgb_to_bw(im_1)
```

```python
plt.imshow(black_and_white_img,cmap='gray')
```
